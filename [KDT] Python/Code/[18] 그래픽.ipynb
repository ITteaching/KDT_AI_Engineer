### 1) 윈도우 생성

* 파이썬은 Tcl/Tk와 그 하위 모듈인 터틀로 그래픽 지원함
* Tcl/Tk  
Tcl(Tool Command Language) 스크립트 언어의 일정  
Tk : 크로스 플랫폼 지원하는 GUI 툴킷
* TkInter : Tcl/Tk를 파이썬에서 사용할 수 있도록 인터페이스 해주는 모듈

* 메인 윈도우 생성 : 타이틀 바, 경계선, 출력 표면

````python
## 위젯 배치 및 이벤트

from tkinter import *

main = Tk()
main.title("Tk Test")
# main.geometry("300*200")
lbl = Label(main, text="Label", font="Arial 20")
lbl.pack()

def appleclick():
    lbl["text"] = "Apple"
    
apple = Button(main, text="Apple", foreground="Red", command=appleclick)
apple.pack()

def orangeclick():
    lbl["text"] = "Orange"
    
orange = Button(main, text="Orange", foreground="Green", command=orangeclick)
orange.pack()

main.mainloop()
````

### 3) 메뉴

* Menu 클래스로 생성
* 윈도우 부착 메뉴 바 생성 : 인수로 부모 윈도우 지정 후 윈도우 config 함수로 이 메뉴 지정
* 팝업 메뉴 : 메뉴 바를 부모로 하여 생성, add_cascade 메서드로 부착
* 구체적인 명령 수행 메뉴 항목 생성 : 팝업 메뉴의 add_command 메서드로 생성

````python
from tkinter import *
import tkinter.messagebox

main = Tk()
main.title("menu")

menubar = Menu(main)
main.config(menu=menubar)
menubar.add_command(label="파일")
menubar.add_command(label="도구")
menubar.add_command(label="종료")

def show():
    tkinter.messagebox.showinfo("확인","팝업 확인되었습니다.")
    
popup = Menu(menubar)
menubar.add_cascade(label="팝업메뉴", menu=popup)
popup.add_command(label="팝업확인", command=show)
popup.add_command(label="팝업닫기", command=quit)

main.mainloop()
````

## 3) 이벤트 처리

* onkeypress(함수, "키") : 키를 누르면 함수를 호출함
* onkeyrelease(함수, "키") : 키를 놓으면 함수를 호출함
* onscreenclick(함수, 버튼) : 마우스 버튼을 누르면 지정한 함수를 호출함, 버튼 생략 시 1번인 왼쪽 버튼 적용
* ontimer(함수, 시간) : 일정 시간 후에 함수 호출, 시간은 1/1000초 단위
* listen() : 키 입력 포커스 설정

````python
import turtle as t

def draw(head, dist):
    t.setheading(head)
    t.forward(dist)
    
def toleft():
    draw(180, 15)
    
def toright():
    draw(0, 15)
    
def toup():
    draw(90, 15)
    
def todown():
    draw(270, 15)
    
def move(x,y):
    t.up()
    t.setpos(x,y)
    t.down()

t.shape("turtle")
t.speed(0)
t.onkeypress(toleft, "Left")
t.onkeypress(toright, "Right")
t.onkeypress(toup, "Up")
t.onkeypress(todown, "Down")
t.onscreenclick(move)
t.listen()
t.done()
````
